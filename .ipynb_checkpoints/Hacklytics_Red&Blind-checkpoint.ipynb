```python
import pandas as pd

df = pd.read_csv('Sandy_Springs_Police_Department - Sandy_Springs_Police_Department.csv')

df.head()
df.shape
df['parent_incident_type'].describe()
```

Prediction: parent_incident_type

```python
df.dtypes  #check datatypes to identify unwanted rows

#need to convert incident_datetime, day_of_week
```

```python
month = []
year = []
for datetime in df.incident_datetime:
    month = month + [(int(datetime[0:2]))]
    year = year + [int(datetime[6:10])]
df['incident_month'] = month
df['incident_year'] = year
```

```python
df['day_of_week'] = df['day_of_week'].map({'Sunday': 0, 'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6}).astype(int)
df.head()
```

```python
df_train = df
df.head()
```

```python
from sklearn import preprocessing #preprocessing is used to normalize the data
from sklearn.model_selection import train_test_split #train test split is used for preprocessing
```

```python
features = ['day_of_week', 'hour_of_day', 'incident_month', 'incident_year', 'latitude', 'longitude']
#day_of_week incident_datetime (split it month and year) hour_of_day longitude latitude
```

```python
X = df_train[features]
y = df_train['parent_incident_type']
```

```python
scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X) #Normalizing the feature training set
X
```

```python
pd.DataFrame(X)
```

```python
#transforming incoming data for evaluation
sample_result = [['6', '14', '4', '2015', '34.5772928', '-83.9594747']]
#'day_of_week', 'hour_of_day', 'incident_month', 'incident_year', 'latitude', 'longitude'

sample_result_transformed = scaler.transform(sample_result)
pd.DataFrame(sample_result_transformed)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn import svm
```

```python
#training
classifier = svm.SVC(kernel='rbf', gamma='auto', C=2)
#kernel available: linear, polynomial, radial basis function
#Large Value of parameter C => small margin
#Small Value of paramerter C => Large margin
classifier.fit(X_train, y_train)
```

```python
y_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_predict))
pd.DataFrame(confusion_matrix(y_test,y_predict))
```
